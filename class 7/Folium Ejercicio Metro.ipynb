# Folium
Ejercicio `estaciones del metro`

```python
import folium
import pandas as pd
```

```python
df =pd.read_csv('estaciones-metro.csv')
df
```

```python
cdmx =[19.43272807826792, -99.13337708943038]
m = folium.Map(location=cdmx, tiles="CartoDB Positron", zoom_start=10)
```

```python
df['geopoint'].iloc[0].split(',')
```

```python
for estacion, coordenadas in zip(df['stop_name'], df['geopoint']):
     folium.Marker(coordenadas.split(','), 
                  popup=estacion, 
                  tooltip='Click me!',
                  icon=folium.Icon(color='green',icon='info-sign')
                 ).add_to(m)
```
