Usar folium para graficar `limite geografico` de cada alcaldia de la `CDMX`

```python
import folium
import pandas as pd
```

```python
df = pd.read_csv("limite-de-las-alcaldias.csv")
df.head()
```

```python
cdmx =[19.43272807826792, -99.13337708943038]
m = folium.Map(location=cdmx, tiles="CartoDB Positron", zoom_start=10)
```

```python
for alcaldia in df['geo_shape']:
    folium.Choropleth(geo_data=alcaldia).add_to(m)
m
```

```python
cdmx =[19.43272807826792, -99.13337708943038]
m2 = folium.Map(location=cdmx, tiles="CartoDB Positron", zoom_start=10)
```

```python
for alcaldia,coordenadas in zip(df['nomgeo'],df['geo_point_2d']):
    folium.Marker(coordenadas.split(','), 
                  popup=alcaldia, 
                  tooltip='Click me!',
                  icon=folium.Icon(color='red',icon='cloud')
                 ).add_to(m2)
m2
```
