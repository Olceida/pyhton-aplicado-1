# APIs

```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base_url = 'http://hp-api.herokuapp.com/api/characters'
```

```python
resultados = requests.get(base_url)
```

```python
resultados.status_code
```

```python
resultados.content
```

```python
resultados.text
```

```python
data = json.loads(resultados.text)
data
```

```python
for personaje in data:
    #X personaje nacio en y
    print(f"{personaje['name']} nacio en {personaje['dateOfBirth']}")
```

```python
personajes = []
for elemento in data:
    personajes.append(elemento['name'])
personajes
```

```python
personajes = [elemento['name'] for elemento in data]
personajes
```

```python
fechas_nacimiento = [elemento['dateOfBirth'] for elemento in data]
fechas_nacimiento
```

```python
casas = [elemento['house'] for elemento in data]
casas
```

```python
diccionario = {
    'Personaje': personajes,
    'Fecha de nacimiento': fechas_nacimiento,
    'Casa': casas
}
```

```python
df = pd.DataFrame(diccionario)
df
```

```python
df[df['Casa']=='Slytherin']
```

```python
conteo = df.groupby('Casa').count()
conteo
```

```python
plt.figure(figsize=(10,10))
plt.pie(conteo['Personaje'], labels=conteo.index.values , autopct='%.2f')
```
