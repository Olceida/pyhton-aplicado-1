```python
import requests
import json
import pandas as pd
```

```python
base_url= "https://datos.cdmx.gob.mx/api/3/action/datastore_search"
resource_id = "ff1d4cbf-5985-45db-b40f-d820ce2b01a2"
q = input('Colonia: ')

params = {
    'resource_id':resource_id,
    'q':q
}
```

```python
resultados = requests.get(base_url, params=params)
resultados
```

```python
cobichos = resultados.json()['result']['records']
```

```python
colonia = [cobicho['colonia'] for cobicho in cobichos]
alcaldia = [cobicho['alcaldia'] for cobicho in cobichos]
total = [cobicho['total'] for cobicho in cobichos]
geo_shape = [cobicho['geo_shape'] for cobicho in cobichos]
geo_point = [cobicho['geo_point'] for cobicho in cobichos]
```

```python
diccionario_Covid = {
    'Colonia':colonia,
    'Alcaldia':alcaldia,
    'Total':total,
    'Geo_Shape':geo_shape,
    'Geo_Point':geo_point
}
```

```python
df = pd.DataFrame(diccionario_Covid)
df
```

```python
df.sort_values('Total', ascending=False)
```
